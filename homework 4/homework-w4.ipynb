**(3) Node $A$ needs to identify risky situations in Lappeenranta when the temperature is above $25$ degrees and the wind is above $5$ m/s based on the sensors' signals. If a risky situation is identified at node $A$ as binary number "1" while normal situations as "0". How would you solve this problem? Plot the output signal from node $A$ and discuss its relation to the input signals from $B$-$E$.**

```python
aux=(airport["Air temperature (degC)"].values>25) & (airport["Wind speed (m/s)"].values>5)
aux2=(lepola["Air temperature (degC)"].values>25) & (lepola["Wind speed (m/s)"].values>5)
aux3=(airport["Air temperature (degC)"].values>25) | (airport["Wind speed (m/s)"].values>5)
aux4=(lepola["Air temperature (degC)"].values>25) | (lepola["Wind speed (m/s)"].values>5)
plt.figure(figsize=(16,6)) 
plt.plot(airport["Air temperature (degC)"].index, aux, color='yellow', marker='.',linestyle='') 
plt.plot(airport["Air temperature (degC)"].index, aux2, color='green', marker='.',linestyle='')
plt.plot(airport["Air temperature (degC)"].index, aux3, color='purple', marker='.',linestyle='') 
plt.plot(airport["Air temperature (degC)"].index, aux4, color='pink', marker='.',linestyle='') 
plt.grid(True) 
plt.show()

# If on any weather station in Lappeenranta temperature>25 or wind>5, we have risky situation and it must be sygnalized witth "1". 
# Also if temperature>25 and wind>5 in the same time, it's even more risky. This situations are visible on the plot. 
# We can say, that in 2018 there wasn't a situation when temp. and wind were risky at the same time at lepola, but it happened at the airport. 
# There were several times when temp. or wind at the airport were risky. Same for lepola.

aux5=(airport["Air temperature (degC)"].values>25) | (airport["Wind speed (m/s)"].values>5) | (lepola["Air temperature (degC)"].values>25) | (lepola["Wind speed (m/s)"].values>5)
plt.figure(figsize=(16,6)) 
plt.plot(airport["Air temperature (degC)"].index, aux5, color='red', marker='.',linestyle='') 
plt.grid(True) 
plt.show()

# In general, plot of aux5 shows all the risky situations, in which temperature>25 or wind>5 on any station. So if any of there four values is 1, the output signal is 1 also 1, because we use operator OR.
```
