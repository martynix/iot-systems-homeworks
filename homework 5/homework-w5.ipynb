**(1) Create two arrays with 10000 samples generated as [Binomial](https://en.wikipedia.org/wiki/Binomial_distribution) random variables via [Numpy](https://docs.scipy.org/doc/numpy/reference/generated/numpy.random.binomial.html) similar to what we have done with Poisson random variables in the tutorial considering the following paramenters of the Binomial function:**

**(a) $n=100$ and $p=0.3$;**

**(b) $n=10000$, and $p=0.6$**


**Make the histogram (empirical) plots related to each vector and compare them with the analytical results with the probability distribution funtion given by:**

**$$\mathrm{Prob}\left[X = k \right] = \binom{n}{k}p^k(1-p)^{n-k},$$ where $k$ is the number of *successes* in $n$ trials.**

**Remember to use this [scipy.stats.binom](https://docs.scipy.org/doc/scipy-0.14.0/reference/generated/scipy.stats.binom.html) in a similar way we did in the tutorial with Poisson.**

```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib import style
from scipy.stats import poisson, binom
style.use('bmh')
import warnings
warnings.filterwarnings('ignore')

ar1 = np.random.binomial(100, 0.3, 10000)
ar2 = np.random.binomial(10000, 0.6, 10000)

n1, p1 = 100, 0.3
n2, p2 = 10000, 0.6

k_values = np.arange(0, max(n1, n2)+1)

probabilities1 = binom.pmf(k_values, n1, p1)
probabilities2 = binom.pmf(k_values, n2, p2)

plt.figure(figsize=(10, 5))
plt.hist(ar1, bins=np.arange(-0.5, n1+1.5, 1), density=True, alpha=0.5, label='Empirical histogram for ar1')
plt.step(k_values, probabilities1, where='post', label='Analytical Probability (ar1)')
plt.xlabel('Number of Successes (k)')
plt.ylabel('Probability')
plt.title('Binomial Distribution Comparison (ar1)')
plt.legend()
plt.show()

plt.figure(figsize=(10, 5))
plt.hist(ar2, bins=np.arange(-0.5, n2+1.5, 1), density=True, alpha=0.5, label='Empirical histogram for ar2')
plt.step(k_values, probabilities2, where='post', label='Analytical Probability (ar2)')
plt.xlabel('Number of Successes (k)')
plt.ylabel('Probability')
plt.title('Binomial Distribution Comparison (ar2)')
plt.legend()
plt.show()
```

**(2) Read Section 2.3 and Chapter 3 from [Network Science](http://networksciencebook.com/) and generate three different kinds of Erdos-Renyi graphs with $N=15$ (fifteen nodes) using [NetworkX](https://networkx.github.io/documentation/stable/reference/generated/networkx.generators.random_graphs.erdos_renyi_graph.html?highlight=erdos#networkx.generators.random_graphs.erdos_renyi_graph). 
The networks shall be (a)  with probability $p=0.2$, (b) with probability $p=0.5$ and (c) with probability $p=0.8$. What are the differences you see if these graphs represent communication networks.**

```python
import networkx as nx
import matplotlib.pyplot as plt

N = 15
probabilities = [0.2, 0.5, 0.8]

for idx, p in enumerate(probabilities):
    G = nx.erdos_renyi_graph(N, p)
    plt.figure(figsize=(8, 6))
    nx.draw(G, with_labels=True, node_color='pink', node_size=500, font_size=10, font_color='black')
    plt.title(f'Erdos-Renyi Graph (p={p})')
    plt.show()
    
    

#This graphs differ from each other with number of connections between nodes. The higher the probability, the more connections.
#In a communication network, if nodes have fewer direct connections, it might be causing delays in communication, since the signal must go through many nodes. 
#But in the third graph, there is a lot of connections, so higher traffic is also not very beneficial for communication network. Too big traffic can make the network overloaded.

```
