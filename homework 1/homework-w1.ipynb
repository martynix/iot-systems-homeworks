**(3) Download the air temperature data of one year from [FMI](http://en.ilmatieteenlaitos.fi/download-observations#!/) at the city/region you currently live. Make four (arbitrary) time series plots for one year, one month, one week and one day. Besides, for each one of these time series, compute the average temperature, and plot it in the respective plot as a constant line.**

*Hint:* Code like in the tutorial notebook.

```python
#pip install -r requirements.txt

import pandas as pd
import matplotlib.pyplot as plt

Temperature = pd.read_csv("temp.csv",dayfirst=True,sep=",", 
                          header=0,decimal=b".",index_col=0,
                          parse_dates= [[0, 1, 2, 3]],usecols=[0,1,2,3,5])

Temperature = Temperature[~Temperature['Air temperature (degC)'].str.contains("-")] ;
Temperature = Temperature['Air temperature (degC)'].astype(float)

Temperature.head()
Temperature.info()

#one day
day = Temperature['2021-1-1']
plt.figure(figsize=(16,6)) 
plt.plot(day,color='blue', marker='x',linestyle=':') 
plt.title("Air temperature in jan 1, 2021, at Lappeenranta Konnunsuo") 
plt.ylabel("Temperature in [°C]") 
plt.grid(True) 
plt.show() 

#one week
week = Temperature['2021-1-2':'2021-1-3']
plt.figure(figsize=(16,6)) 
plt.plot(week,color='blue', marker='x',linestyle=':') 
plt.title("Air temperature from jan 1, 2021, to jan 7, 2021, at Lappeenranta Konnunsuo") 
plt.ylabel("Temperature in [°C]") 
plt.grid(True) 
plt.show() 

#one month
month = Temperature['2021-1']
plt.figure(figsize=(16,6)) 
plt.plot(month,color='blue', marker='x',linestyle=':') 
plt.title("Air temperature in jan 2021, at Lappeenranta Konnunsuo") 
plt.ylabel("Temperature in [°C]") 
plt.grid(True) 
plt.show() 

```
